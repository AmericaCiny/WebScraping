## Q2
Write a script that goes to bestbuy.com, clicks on Deal of the Day, reads how much time is left for the Deal of the Day and prints the remaining time to screen (console), clicks on the Deal of the Day (the actual product), clicks on its reviews, and saves the resulting HTML to your local hard drive as "bestbuy_deal_of_the_day.htm".

```python
driver = webdriver.Chrome(executable_path='/Users/xbn/Desktop/Data Design & Representation/HW/4/chromedriver')

driver.set_script_timeout(100)
driver.set_page_load_timeout(100)
time.sleep(3)
driver.get('https://bestbuy.com/')
```

```python
time.sleep(3)
Deal = driver.find_element(By.LINK_TEXT,'Deal of the Day')
Deal.click()
```

```python
time.sleep(3)
time_remaining = driver.find_element(By.ID, 'countdown')
time_remaining_text = time_remaining.text
time_remaining_text = time_remaining_text.replace('\n',' ')
print(time_remaining_text)
```

```python
time.sleep(3)
deal_of_day = driver.find_element(By.ID, 'wf-offer-ad249565-489a-429e-8dc2-5f735cc25108-product-title')
deal_of_day.click()
```

```python
time.sleep(3)
review = driver.find_element(By.LINK_TEXT, 'See All Customer Reviews')
review.click()
```

```python
time.sleep(3)
with open("bestbuy_deal_of_the_day.htm", "w", encoding="utf-8") as f:
        f.write(driver.page_source)
```

## 3
Since you all love Lavazza coffee:  Write a script that goes to lavazzausa.com, rejects all cookies, then clicks on "Products", then on "Capsule coffee machines", save the resulting page to the file "lavazza.htm", and close the browser.  Subsequently write code that opens the saved "lavazza.htm" file and find the original price of the "Classy Plus Bundle" and prints it to screen (please use BeautifulSoup for this).

Bonus:  Given the price tag and the fact that Rachel really loves and cares for our Lavazza machine, please write an apology email to Rachel in case you have ever treated MSBA's Lavazza Classy Plus coffee machine with anything but love and respect.

```python

driver = webdriver.Chrome(executable_path='/Users/xbn/Desktop/Data Design & Representation/HW/4/chromedriver')
driver.implicitly_wait(10)
driver.get("https://www.lavazzausa.com/en")
```

```python
time.sleep(3)
reject = driver.find_element(By.ID, 'consent_prompt_reject')
reject.click()
```

```python
time.sleep(3)

products = driver.find_element(By.ID, 'PRODUCTS')
ActionChains(driver).move_to_element(products).perform()
products.click()
```

```python
time.sleep(3)
capsule_cofee_machine = driver.find_element(By.LINK_TEXT, 'Capsule coffee machines')
capsule_cofee_machine.click()
```

```python
time.sleep(3)
with open("lavazza.htm", "w", encoding="utf-8") as f:
        f.write(driver.page_source)
driver.quit()
```

```python
with open('lavazza.htm', 'r', encoding='utf-8') as file:
    htm = file.read()
soup = BeautifulSoup(htm, 'html.parser')
print(soup.prettify())
```

```python
price = soup.select_one('article[data-product-name="Classy Plus"]')['data-price']
print("Price of the Classy Plus is: $", price)
```
